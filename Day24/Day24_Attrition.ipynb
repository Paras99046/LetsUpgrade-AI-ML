```python
new_com_worked = np.where(data['NumCompaniesWorked'].isnull(),3,data['NumCompaniesWorked'])
new_total_working_years = np.where(data['TotalWorkingYears'].isnull(),11,data['TotalWorkingYears'])
```

```python
data['NumCompaniesWorked'] = new_com_worked
data["TotalWorkingYears"] = new_total_working_years
```

```python
data.isnull().sum()
```

```python
from sklearn import preprocessing
```

```python
label_encoder = preprocessing.LabelEncoder()
```

```python
encoded_attrition = label_encoder.fit_transform(data['Attrition'])
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
data.columns
```

```python
rf_model = RandomForestClassifier(n_estimators = 1000,max_features = 2,oob_score = True)
```

```python
features = ['Age','DistanceFromHome','Education','MonthlyIncome', 'NumCompaniesWorked', 'PercentSalaryHike','TotalWorkingYears',
            'TrainingTimesLastYear', 'YearsAtCompany','YearsSinceLastPromotion', 'YearsWithCurrManager']
```

```python
rf_model.fit(X=data[features],y=encoded_attrition)
```

```python
rf_model.oob_score_
```

```python
for feature,imp in zip(features,rf_model.feature_importances_):
    print(feature,imp)
```

```python
from sklearn import tree
```

```python
tree_model = tree.DecisionTreeClassifier(max_depth = 8)
```

```python
predictors = pd.DataFrame([data['MonthlyIncome'],data['Age'],data['TotalWorkingYears']]).T
```

```python
tree_model.fit(X = predictors,y=data['Attrition'])
```

```python
print(tree_model.score(X = predictors,y = data['Attrition']))
```

```python
with open('DTree_Attrition-3var.dot','w') as f:
    f = tree.export_graphviz(tree_model,feature_names = ['MonthlyIncome','Age','TotalWorkingYears'],out_file=f)
```
