```python
import numpy as np
import pandas as pd
```

```python
filepath = r"C:\Users\admin\Documents\LetsUpgrade\Notes\Day 21\Bank_Personal_Loan_Modelling.xlsx"
excel = pd.ExcelFile(filepath)
```

```python
excel.sheet_names
```

```python
data = excel.parse('Data')
data.head()
```

```python
data.isnull().sum()
```

```python
data = data.drop('ZIP Code',axis = 1)
data
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
data.columns
```

```python
rf_model = RandomForestClassifier(n_estimators = 1000,max_features = 2,oob_score = True)
```

```python
features = ['ID', 'Age', 'Experience', 'Income', 'Family', 'CCAvg', 'Education',
       'Mortgage', 'Securities Account', 'CD Account',
       'Online', 'CreditCard']
```

```python
rf_model.fit(X=data[features],y=data['Personal Loan'])
```

```python
rf_model.oob_score_
```

```python
for feature,imp in zip(features,rf_model.feature_importances_):
    print(feature,imp)
```

```python
from sklearn import tree
```

```python
tree_model = tree.DecisionTreeClassifier(max_depth = 8)
```

```python
predictors = pd.DataFrame([data['Income'],data['CCAvg'],data['Education']]).T
```

```python
tree_model.fit(X = predictors,y=data['Personal Loan'])
```

```python
print(tree_model.score(X = predictors,y = data['Personal Loan']))
```

```python
with open('DTree_Bank_Loan_Modelling.dot','w') as f:
    f = tree.export_graphviz(tree_model,feature_names = ['Income','CCAvg','Education'],out_file=f)
```
